```python
idx1 = df['nombre_act'].values == 'Escuelas de educación primaria del sector privado'
idx2 = df['nombre_act'].values == 'Escuelas de educación primaria del sector público'

plt.figure()
plt.subplot(1,3,1)
plt.title('Aguascalientes')
idxA = df['municipio'].values =='Aguascalientes'
c1A = np.sum( np.logical_and(idx1,idxA) ) #privado
c2A = np.sum( np.logical_and(idx2,idxA) ) #público
plt.pie( [c1A,c2A], labels=['escuelas primarias privadas', 'escuelas primarias públicas'] )

plt.subplot(1,3,2)
plt.title('Calvillo')
idxC = df['municipio'].values =='Calvillo'
c1C = np.sum( np.logical_and(idx1,idxC) ) #privado
c2C = np.sum( np.logical_and(idx2,idxC) ) #público
plt.pie( [c1C,c2C], labels=['escuelas primarias privadas', 'escuelas primarias públicas'] )

plt.subplot(1,3,3)
plt.title('Cosío')
idxCs = df['municipio'].values =='Cosío'
c1Cs = np.sum( np.logical_and(idx1,idxCs) ) #privado
c2Cs = np.sum( np.logical_and(idx2,idxCs) ) #público
plt.pie( [c1Cs,c2Cs], labels=['escuelas primarias privadas', 'escuelas primarias públicas'] )
plt.show()
```
