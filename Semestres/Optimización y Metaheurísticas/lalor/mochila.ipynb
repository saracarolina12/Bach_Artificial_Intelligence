```python
def mochila(pesos,ganancias,peso_max,K=1000,T=1000,e=1e-8,alfa=0.99,penalizacion=1000000):
    def crear_vector():
        vec=[]
        for i in range(len(ganancias)):
            vec.append(1 if np.random.uniform() > 0.99 else 0)
        return vec
    def modificar_vector(vec):
        pos=np.random.randint(0,len(vec)-1)
        vec[pos]= int(not vec[pos])
        return vec
    def obtener_peso(vec):
        peso_total=sum(np.array(vec)*np.array(pesos))
        return peso_total
    def obtener_ganancia(vec):
        total=sum(np.array(vec)*np.array(ganancias))
        return (total-penalizacion if obtener_peso(vec)>peso_max else total)
    k=0
    soluciones=[[],[],[]]
    soluciones[0].append(crear_vector())
    soluciones[1].append(obtener_peso(soluciones[0][-1]))
    soluciones[2].append(obtener_ganancia(soluciones[0][-1]))
    mejor_solucion=0
    print(soluciones[0][-1])
    while(k<K and T>e):
        x_casita=modificar_vector(soluciones[0][-1])
        if(obtener_ganancia(x_casita)>obtener_ganancia(soluciones[0][-1])):
            soluciones[0].append(x_casita)
            soluciones[1].append(obtener_peso(x_casita))
            soluciones[2].append(obtener_ganancia(x_casita))
        else:
            r=np.random.uniform()
            if(r<np.exp((obtener_ganancia(soluciones[0][-1])-obtener_ganancia(x_casita))/T)):
                soluciones[0].append(x_casita)
                soluciones[1].append(obtener_peso(x_casita))
                soluciones[2].append(obtener_ganancia(x_casita))
        if(mejor_solucion<obtener_ganancia(x_casita) and obtener_peso(x_casita)<=peso_max):
            print(x_casita,obtener_ganancia(x_casita),obtener_peso(x_casita))
            mejor_solucion=obtener_ganancia(x_casita)
        T=alfa*T
        k=k+1

    return soluciones[0][-1], obtener_peso(soluciones[1][-1]), obtener_ganancia(soluciones[2][-1]), soluciones
    
```
