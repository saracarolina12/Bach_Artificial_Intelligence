```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
import cv2
```

```python
img = cv2.imread('../datasets/doggie.jpg')
img
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convertir color [viene en cv]
plt.figure()
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
nrows, ncols, nch = img.shape
vect = np.reshape(img,(nrows*ncols,3))
copy = np.reshape(img,(nrows*ncols,3))
```

### KMeans

```python
model = KMeans(n_clusters=3)
g = model.fit_predict(vect)

print(g)
```

```python
#print(g.size)
ngroups = np.unique(g)
H = np.zeros(ngroups.size)
S = np.zeros(ngroups.size)
V = np.zeros(ngroups.size)
# print(ngroups.size) #0 1 2
print(copy)
for i in range(0, ngroups.size):
    H[i] = np.mean(vect[g==i,0])
    S[i] = np.mean(vect[g==i,1])
    V[i] = np.mean(vect[g==i,2])
    copy[g==i,0] = H[i]
    copy[g==i,1] = S[i]
    copy[g==i,2] = V[i]
    
final = np.reshape(copy,(nrows,ncols,nch))   

plt.figure()
plt.imshow(final)
plt.show()
```

### Gaussian Mixture

```python
ROWS, COLS, NCH = img.shape
myvect = np.reshape(img,(ROWS*COLS,3))
mycopy = np.reshape(img,(ROWS*COLS,3))
```
