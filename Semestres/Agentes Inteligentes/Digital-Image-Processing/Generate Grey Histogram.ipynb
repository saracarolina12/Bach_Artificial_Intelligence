```python
import numpy as np
import skimage.color
import skimage.io
import matplotlib.pyplot as plt


# read the image of a plant seedling as grayscale from the outset
image = skimage.io.imread(fname="../images/cityscapes.jpg", as_gray=True)

# display the image
fig, ax = plt.subplots()
plt.imshow(image, cmap="gray")
plt.show()
n = image.shape[0]
m = image.shape[1]

values = []
for i in range(0, n-1):
    for j in range(0, m-1):
        values.append(image[i][j])
#         print(image[i][j])
        
n_bins = n
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(values,bins = n_bins)
# Show plot
plt.show()
```
