```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('../../data/datos.csv')
X = df['x'].values
Y = df['y'].values
plt.figure()
plt.plot(X,Y,'.')
plt.show()
```

```python
def objective_function(P,X,Y):
    A,B,C = P[0],P[1],P[2]
    Yp = A*np.sin(B*X+C*np.pi)
    return np.sum(np.abs(Y-Yp))

def draw_solution(P,X,Y):
    A,B,C = P[0],P[1],P[2]
    Yp = A*np.sin(B*X+C*np.pi)
    plt.figure()
    plt.plot(X,Y,'.')
    plt.plot(X,Yp)
    plt.show()

P = [10,10,10]
print(objective_function(P,X,Y))
draw_solution(P,X,Y)
```

# Scipy Differential Evolution

```python
from scipy.optimize import differential_evolution
```

```python
# P = [Z,B,C]
# f(x) = A*sin(Bx+C*pi)
bounds = [  [-100,100], # rango A
            [-100,100], # rango B
            [-100,100]] # rango C
```

```python
res = differential_evolution(objective_function,bounds, args=(X,Y))
res
```

```python
from matplotlib.pyplot import draw


P = res.x
print(objective_function(P,X,Y))
draw_solution(P,X,Y)
```
